18.What precisely is the train-dev kit, when will you need it, how do you put it to use?
While working with machin elearning we can face majorly 3 problems post training overfitting,underfitting and data mismatch. The validation set and test set should be represenattive of training dataset. But there could be situation when they don not represent each other. 
Suppose we have trained a model using training dataset. Post training to check perform,ance measure we use validation dataset, there could be 2 situations now. Model will give perform good wiht validation dataset or it will not. If it doesnt perform well with validation dataset we cans ay that either the model is overfitting or theere is data mismatch problem which means the validation dataset is not representative of training dataset. But right now we do not have any ways to check the exact problem.

While training we can keep a portion of the training dataset separate which is called as train-dev set, and post training we can check the performance of the model on train-dev set and validation set both. If the model performs well with train-dev set and poor with validation set then we can say model is not overfitting but the validation dataset is clearly not the representative of the training dataset. In this case we need to change our dataset so that they can be representative of each other.
