3. What is machine learning, and how does it work? What are the key responsibilities of machine learning?
MAchine learning utilizes whole lot of data to uncover/learn the hidden pattern in the dataset. Most of the algorithms in Machine Learning works on equation and tries to find the value of the parameters using the dataset. It tries to minimize the difference between actual and predicted and thus on reaching the global minima it acquires the optimum value. These parameters are tested on train-dev set and validation dataset to ensure if there is any dataset mismatch or overfitting taking place. If yes these parameters are modified again either by taking more accurate values or by tuning the hyper parameters. If there is not any mismatch/overfitting/underfitting the training parameters are checked with test dataset. BAsed on the problem the performance measure are checked and ensured to achieve less error.

10. What is classification, exactly? What are the main distinctions between classification and regression?
Classification is a task of predicting the class for any input set of features based on probability value rather than predicting any Continous numerical value. Log of odds are calculated, sigmoid function on logg of odds is applied to restrict the output between 0 and 1 (binary) . A threshold is decided to differentiate between two classes depending upon the nature of problem.
The performance measures are based on true classification and false classifcation.  Sometimes accuracy(percentage of true classifcation) doesnt decide how good or bad a model is. Performance measures such as Recall, precision, F1score decides the reliability of the model. Classification Model reliability is problem specific. 
For example: Detecting cancer causing cell Problem want recall performance measure o be high rather than precision which means it is ok to receive non cancer causing cells present in the body (whenn it is actually absent) rather than compromising result when the prediction says that cancer causing cellss are abset when in real it is present. 
This is not case with regression Problem.

Task of predicting continous numerical value  for any input set of features when there is no correlation between input set of features(multicollinearity), one or more feature capturing the most of the variance of target column. The cost function of the regression is a convex function which assumes only one global minima and this is the reasion gRadient descent works well with Cost function of Regreesion. Cost function is differentiated wrt different parameters and new value of the parameters are calculated at each small step. The steps are kept small in ordfer to reach the global minima and the negative/positive slope decides the movement of steps in gradient descent.
